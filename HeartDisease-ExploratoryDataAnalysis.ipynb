IMPORTING LIBRARIES AND CSV

```python
import numpy as np 
import pandas as pd
import seaborn as sns
df=pd.read_csv("heart.csv")
df.head()
```

DISPLOT

```python
sns.displot(df["Age"],kind="hist",rug="True",kde="true")
```

```python
sns.displot(df['MaxHR'], kde = True, color = 'blue')
```

```python
sns.displot(x = df['RestingBP'], kde = True, color = 'red')
```

```python
sns.displot(df['Cholesterol'], kde = True, color = 'pink')
```

```python
df.groupby("Sex").size().plot(kind="pie",autopct="%1.1f%%",figsize=(4,6),title="Sex")
```

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.1f', figsize = (4,6),title="ChestPainType")
```

```python
df.groupby('RestingECG').size().plot(kind = 'pie', autopct = '%.1f', figsize = (4,6),title="RestingECG")
```

```python
df.groupby('ST_Slope').size().plot(kind = 'pie', autopct = '%.1f', figsize = (4,6),title="ST_Slope")
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1f', figsize = (4,6),title="HeartDisease")
```

VIOLIN PLOT

```python
sns.violinplot(x="Age",data=df)
```

```python
sns.violinplot(x="HeartDisease",y="Sex",data=df)
```

```python
sns.violinplot(x = 'HeartDisease',y = 'Age',data=df)
```

```python
sns.violinplot(x="HeartDisease",y="Cholesterol",data=df)
```
